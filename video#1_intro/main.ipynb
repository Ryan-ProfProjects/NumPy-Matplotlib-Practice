```python
x = np.linspace(0, 10, 1000)
y = np.linspace(0, 10, 1000)
f = lambda x, y: x*np.cos(x) + y*np.sin(y)
f(x, y)
```

```python
import matplotlib.pyplot as plt
```

```python
X, Y = np.meshgrid(x, y)
```

```python
x.shape, y.shape
```

```python
X.shape, Y.shape
```

```python
plt.contourf(X, Y, f(X, Y))
plt.colorbar()
```

```python
plt.plot(X, Y, f(X, Y))
```

```python
ax = plt.subplot(projection = "3d")
ax.plot_surface(X, Y, f(X, Y))
```
