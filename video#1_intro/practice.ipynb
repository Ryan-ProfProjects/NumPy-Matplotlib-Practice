```python
# Finding critical points:
# if positive to negative or negative to positive, the derivative at that point is 0, so it is a critical point
```

```python
dydx
```

```python
signs = []
for i in range(len(dydx)):
    if dydx[i] < 0:
        signs.append("negative")
    if dydx[i] > 0:
        signs.append("positive")
signs
```

```python
critical_points = 0
for i in range(len(signs)-1):
    if signs[i] == "negative" and signs[i+1] == "positive":
        print(signs[i], signs[i+1])
        critical_points+=1
    if signs[i] == "positive" and signs[i+1] == "negative":
        print(signs[i], signs[i+1])
        critical_points+=1

critical_points
```

```python
# we only proved there are 3 critical points but not what they are!!!
```

```python
dydx[1:] # skips first element but has last element
```

```python
dydx[:-1] # includes the first element but excludes last element (so these are the first points)
```

```python
dydx[:-1] * dydx[1:] < 0 # now will give boolean array
```

```python
# Polar coordinates time now!
# Consider the flower petal r = 1+(0.75)*sin(3*x) for [0, 2*pi], make a plot of it!
# Compute the area using A = integral from 0 to 2*pi of (0.5)*r**2 * dx
# Compute the arclength using A = integral from 0 to 2*pi of sqrt(r**2 + (dr/dx)^2) * dx
```

```python
2*np.pi
```

```python
theta = np.linspace(0, 2*np.pi, 1000)
theta
```

```python
r = 1+(0.75)*np.sin(3*theta)
r
```

```python
plt.polar(theta, r)
```

```python
# ... or convert from polar to cartesian coordinates:
y = r*np.sin(theta)
x = r*np.cos(theta)
plt.plot(x, y)
```

```python
A = 1/2 * sum(r**2) * (theta[1]-theta[0])
A
```

```python
# Now compute the arclength using L = integral from 0 to 2*pi of sqrt(r**2 + (dr/dx)^2) * dx
drdtheta = np.gradient(r, theta)
# theta[n] - theta[n-1] where n is within the shape of the array, is the width eleemnt, dtheta or delta theta
L = sum(np.sqrt(r**2 + drdtheta**2)) * (theta[1]-theta[0]) # riemann sum
L
```

```python
integrand = np.sqrt(r**2 + drdtheta**2)
np.trapz(integrand,theta) # trapeziodal sum
```

```python
# dealing with 2D functions now!!
# you now have to use meshgrids to have 2D functions. Or then it will have shape errors
x = np.linspace(0, 10, 1000)
y = np.linspace(0, 10, 1000)
```

```python
xv, yv = np.meshgrid(x, y)
xv
```

```python
zv = xv**2 + yv**2
plt.contourf(xv, yv, zv, levels=30)
plt.colorbar()
```

```python
f = lambda x, y: x**2 + y**2
z = f(xv, yv)
dzdx, dzdy = np.gradient(z, x, y)
plt.contourf(x, y, dzdx, levels=30)
plt.colorbar()
```

```python
plt.contourf(x, y, dzdy, levels=30)
plt.colorbar()
```

```python
ax = plt.subplot(projection="3d")
ax.plot_surface(x, y, z, cmap="viridis")
plt.show()
```
